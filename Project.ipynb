```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import pandas as pd
import time

driver = webdriver.Chrome()
driver.get('https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=fantasy')
driver.maximize_window()
driver.execute_script("""var ads = document.querySelectorAll('iframe, .adsbygoogle, .ad-banner');
        ads.forEach(ad => ad.style.display = 'none');""")

def click_loadbutton():
    try:
        load_button = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/div[2]/div/span/button/span')
        ActionChains(driver).move_to_element(load_button).perform()
        load_button.click()
        time.sleep(3)
        return True
    except Exception as e:
        print("Error")
        return False

while click_loadbutton():
    print("click")

print("not clicked")

# Scrape data
movie = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li/div/div/div/div[1]/div[2]/div[1]/a/h3')
duration = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li/div/div/div/div[1]/div[2]/div[2]/span[2]')
rating = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li/div/div/div/div[1]/div[2]/span/div/span/span[1]')
voting = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li/div/div/div/div[1]/div[2]/span/div/span/span[2]')

# Initialize empty lists
m = []
d = []
r = []
v = []

# Scrape movie titles
for i in movie:
    s=i.text if i.text else None
    st=""
    for j in range(len(s)):
        if(((s[j]>='0'and s[j]<='9') and j>=0 and j<=2)or s[j]=='.' ):
            continue
        elif((s[j]>='0'and s[j]<='9')):
            st=st+s[j]
        else:
            st=st+s[j]
    m.append(st)

for j in duration:
    dur=j.text if j.text else None
    d.append(dur)
for k in rating:
    rat=k.text if k.text else None
    r.append(rat)
for o in voting:
    cc=-1
    s=o.text if o.text else None
    s=s[2:len(s)-1]
    s1=""
    decimal=0.0
    for i in range(len(s)):
        if(s[i]=='K' or s[i]=='k'):
            decimal=float(s1)
            cc=1
            decimal=decimal*1000
        else:
            s1=s1+s[i]
    if(cc==1):
        v.append(decimal)
    else:
        decimal=float(s1)*1000
        v.append(decimal)

# Ensure all lists have the same length by filling missing data
max_len = max(len(m), len(d), len(r), len(v))

# Extend the shorter lists with None
m.extend([None] * (max_len - len(m)))
d.extend([None] * (max_len - len(d)))
r.extend([None] * (max_len - len(r)))
v.extend([None] * (max_len - len(v)))

# Create DataFrame
df = pd.DataFrame({
    'Title': m,
    'Duration': d,
    'Rating': r,
    'Voting': v
})

# Save to CSV
df.to_csv(r'C:\Users\kar81241\OneDrive - Viavi Solutions Inc\Documents\AIML prjct\fantasy_movies.csv', index=False)
```

```python
import pandas as pd
import os

csv_files = [
    r'C:\Users\kar81241\OneDrive - Viavi Solutions Inc\Documents\AIML prjct\adventure_movies.csv',  
    r'C:\Users\kar81241\OneDrive - Viavi Solutions Inc\Documents\AIML prjct\fantasy_movies.csv',  
    r'C:\Users\kar81241\OneDrive - Viavi Solutions Inc\Documents\AIML prjct\comedy_movies.csv',   
    r'C:\Users\kar81241\OneDrive - Viavi Solutions Inc\Documents\AIML prjct\documentary_movies.csv',
    r'C:\Users\kar81241\OneDrive - Viavi Solutions Inc\Documents\AIML prjct\romance_movies.csv'
]

dfs = []

for genre_file in csv_files:
    genre_df = pd.read_csv(genre_file)
    genre_name = os.path.basename(genre_file).split('_')[0].capitalize()
    genre_df['Genre'] = genre_name
    
    dfs.append(genre_df)

combined_df = pd.concat(dfs, ignore_index=True)
combined_df = combined_df[['Title', 'Duration', 'Rating', 'Voting', 'Genre']]

combined_df.to_csv(r'C:\Users\kar81241\OneDrive - Viavi Solutions Inc\Documents\AIML prjct\IMDB_movies.csv', index=False)

print("CSV files combined successfully!")
```
